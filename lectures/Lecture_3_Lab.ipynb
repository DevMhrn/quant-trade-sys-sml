```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

# Simulate prices
prices = 100 + np.random.normal(0.2, 1, 200).cumsum()

# Compute returns
returns = pd.Series(prices).pct_change()

# Rolling volatility
volatility = returns.rolling(20).std()

# Moving averages
ma20 = pd.Series(prices).rolling(20).mean()
ma50 = pd.Series(prices).rolling(50).mean()
```

```python
# Plot prices
plt.figure(figsize=(10,5))
plt.plot(prices)
plt.title('Price Series')
plt.show()
```

```python
# Plot returns
plt.figure(figsize=(10,5))
plt.plot(returns)
plt.title('Returns')
plt.show()
```

```python
# Plot volatility
plt.figure(figsize=(10,5))
plt.plot(volatility)
plt.title('Rolling Volatility')
plt.show()
```

```python
# Plot moving averages
plt.figure(figsize=(10,5))
plt.plot(prices, label='Price')
plt.plot(ma20, label='20 MA')
plt.plot(ma50, label='50 MA')
plt.legend()
plt.title('Moving Averages')
plt.show()
```
