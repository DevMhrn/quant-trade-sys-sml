```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reproducibility
np.random.seed(42)
```

```python
# Simulate bid prices
bid = 100 + np.random.normal(0, 0.2, 100).cumsum()
```

```python
# Ask prices slightly higher
ask = bid + 0.5
```

```python
# Compute spread
spread = ask - bid
```

```python
# Market order execution
market_buy_price = ask
market_sell_price = bid
```

```python
# Limit order example
limit_buy_price = 99.5
```

```python
# Create dataframe
df = pd.DataFrame({
    'Bid': bid,
    'Ask': ask,
    'Spread': spread
})
```

```python
print(df.head())
```

```python
# Average spread
print("Average Spread:", spread.mean())
```

```python
# Plot bid/ask
plt.figure(figsize=(10,5))

plt.plot(bid, label='Bid')
plt.plot(ask, label='Ask')

plt.title("Bid vs Ask Prices")
plt.legend()

plt.show()
```

```python
# Plot spread
plt.figure(figsize=(10,5))

plt.plot(spread)

plt.title("Spread Over Time")

plt.show()
```
