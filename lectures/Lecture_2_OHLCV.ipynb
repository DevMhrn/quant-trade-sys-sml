```python
import numpy as np
import pandas as pd

np.random.seed(42)

open_price = 100 + np.random.normal(0,1,10).cumsum()
high = open_price + np.random.uniform(0.5,2,10)
low = open_price - np.random.uniform(0.5,2,10)
close = open_price + np.random.normal(0,1,10)
volume = np.random.randint(1000,10000,10)

df = pd.DataFrame({
    'Open': open_price,
    'High': high,
    'Low': low,
    'Close': close,
    'Volume': volume
})

print(df)
print(df["Close"] - df["Open"])
```
